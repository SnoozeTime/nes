```python
# f = CPU / (16 * (t + 1))
# t = (CPU / (16 * f)) - 1
fcpu = 1.789773*1e6
def freq_from_timer(t):
    return fcpu / (16 * (t+1))
def timer_from_freq(f):
    return (fcpu / (16 * f)) - 1
```

```python
def nb_samples_up(desired_f, duty=0.25): 
    fcpu = 1.789773 * 1000000 
    cpu_per_frame = fcpu/60 
    per_frame = desired_f / 60 
    return (cpu_per_frame/per_frame) * duty 
```

```python
nb_samples_up(380, 0.125)
```
